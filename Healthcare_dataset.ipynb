**Want to know number of days patient spent at the hospital**

```python
df['Blood Group'] = np.nan
for i in range(len(df['Blood Group'])):
  df.loc[i, 'Blood Group'] = df.loc[i, 'Blood Type'][:-1]
```

```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
df['Duration'] = np.nan
for i in range(len(df['Duration'])):
     df.loc[i, 'Duration'] = (df.loc[i, 'Discharge Date'] - df.loc[i, 'Date of Admission']).days


df['Duration'] = df['Duration'].astype(int)
```

```python
df.head()
```

**I drop name and room number**

```python
df.drop(['Name', 'Room Number'], axis = 1, inplace = True)
```

```python
for col in df.columns:
    if df[col].dtype == 'O' and col != 'Doctor' and col!= 'Hospital':
        print(df[col].value_counts())
        print('--')
```
